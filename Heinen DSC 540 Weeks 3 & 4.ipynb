```python
# Initialise list to store names in 
names = []

# Open the text file. as f ( we are using a loop to go through every line. 
with open('adult_income_names.txt','r') as f:
    # Loop by line of the text file. 
    for line in f:
        # read each line 
        f.readline()
        # assign a variable value using the ":" as the split or deliniator. 
        var=line.split(":")[0]
        # add the variable to the names list 
        names.append(var)
# Display names list 
names
```

```python
# 8) Create a function to strip the whitespace charachters. 

# Define Function 
def strip_whitespace(s):
    # Use strip method to return a string with leading and trailing whitespace removed.
    return s.strip()
```

```python
# 9) apply the strip_whitespace function to all columns containing string values 
# create an new column 
# Copy values from new column to old column 
# drop the new column 

# Use apply to education column 
df_subset['education_stripped'] = df['education'].apply(strip_whitespace)
# copy new values to old column. 
df_subset['education'] = df_subset['education_stripped']
# Use drop method to drop the new (temp column) 
df_subset.drop(labels = ['education_stripped'],axis=1,inplace=True)

# Use apply to Occupation column
df_subset['occupation_stripped'] = df['occupation'].apply(strip_whitespace)
# copy new values to old column. 
df_subset['occupation']  =df_subset['occupation_stripped']
# Use drop method to drop the new (temp column) 
df_subset.drop(labels = ['occupation_stripped'],axis=1,inplace=True)

# Validate
df_subset.head()
```

```python
# 10) Find the number of people who are aged between 30 and 50 

# Use the & clause to create a dataset containing only ages above 30 and below 50. 
# This method will combine the results of two boolean conditions
df_filtered=df_subset[(df_subset['age']>=30) & (df_subset['age']<=50)]
```

```python
# Import packages 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# 2) reas the page using bs4
# open the HTML using encoding UTF-8
fd = open("List of countries by GDP (nominal) - Wikipedia.htm", "r", encoding = "utf-8")
# Use beautiful soup to parse the page and find tables. 
soup = BeautifulSoup(fd)
# Close HTMl page
fd.close()
```

```python
# 3) find the table structure you will need to deal with (How many tables are there?

# Used find all tables to determine the number of tables in the HTML
num_tables = soup.find_all("table")

# use legth element to print the number of tables 
print("Total number of tables are: ", (len(num_tables)))
```

```python
# 4) find the right table using bs4 

# Use soup. find method to find the first element with the class of wikitable. (
# these are the tables we want) 
# Assign that table to the variable data_table
data_table = soup.find("table", {"class": '"wikitable"|}'})

# Print the type of the data table 
print(type(data_table))

# View data table so we know how to parse. 
data_table

```

```python
# isolate the three tables we are looking for, the 3 tables are denoted by 'tr'
# use find all method to find all three sources. 
sources = data_table.tbody.findAll('tr', recursive=False)[0]

# Create sources list variable. 
# itterate through sources list to find number of td elements. 
# td seperates each table

sources_list = [td for td in sources.findAll('td')]
# Use length method to print the number of tables in sources list. 

print('Number of tables in sources_list: ' , len(sources_list))

# display
sources_list
```

```python
# Use find all to find all tr elements within table body, then select the second row 
# use indec 1 
# then create a list of all td elements withing that row and assign to data variable 
data = data_table.tbody.findAll('tr', recursive=False)[1].findAll('td', recursive=False)

# display
data
```

```python
# initialise data_tables list 
data_tables = []

# loop through data and for every td element in list...
# append it to the list of data tables
# this loop is trying to find only direct child elements. 
for td in data:
    data_tables.append(td.findAll('table'))
# verify
print('number of data tables: ', len(data_tables))
```

```python
# 6) get source names from the list of sources you have created . 

# retreive source names for evry table. 

# loop through every source in source list and retreive the source names. 
# the source names are denoted by the anchor <a>. 
source_names = [source.findAll('a')[0].getText() for source in sources_list]

# print 
print(source_names)
```

```python
# 7) Seperate the header and data from the data seperated for 1st souce only create a dataframe. 

# Find the headers in data tables (only the first one (use index 0) 
# use gettext and strip methods to find each thread element in the 1 st table. 
header1 = [th.getText().strip() for th in data_tables[0][0].findAll('thead')[0].findAll('th')]
# print headers. 
header1
```

```python
# In the first data tablke find all body elements and find all tr elements and assign to rows 1 
rows1 = data_tables[0][0].findAll('tbody')[0].findAll('tr')
```

```python
#find all of the td elements in rows 1 and assign to data_rows1
# use get text and strip to retreve and clean whitespace. 
data_rows1 = [[td.get_text().strip() for td in tr.findAll('td')] for tr in rows1]
```

```python
# Build data frme with data from data_rows1 and column names from header 1
df1 = pd.DataFrame(data_rows1, columns=header1)
# display 
print(source_names[0])
df1.head()
```

```python
# 8) repeat for other 2 data sources. 

# Find the headers in data tables (only the 2nd one (use index 1) 
# use gettext and strip methods to find each thread element in the 1 st table. 
header2 = [th.getText().strip() for th in data_tables[1][0].findAll('thead')[0].findAll('th')]
# print headers. 
header2
```

```python
# In the 2nd data tablke find all body elements and find all tr elements and assign to rows 2 
rows2 = data_tables[1][0].findAll('tbody')[0].findAll('tr')
```

```python
# Build a function to find right text 

def find_right_text(i, td):
    # When the index  is 0 use get text and strip to return data
    if i == 0:
        return td.getText().strip()
   # When the index  is 1 use get text and strip to return data
    elif i == 1:
        return td.getText().strip()
   # if neither of the above conditions is met. 
    else:
        # return the stripped text from the element after the spade charachter. 
        index = td.text.find("")
        return td.text[index+1:].strip()
```

```python
#find all of the td elements in rows 2 and assign to data_rows2
# itterate through each tr element in rows2 
#for each td element found call find the right text passsing and index an the td element. 
data_rows2 = [[find_right_text(i, td) for i, td in enumerate(tr.findAll('td'))] for tr in rows2]
```

```python
# Build data frme with data from data_rows2 and column names from header 2
df2 = pd.DataFrame(data_rows2, columns=header2)
# display source name and table 
# use index 1 
print(source_names[1])
df2.head()
```

```python
# Find the headers in data tables (only the 3nd one (use index 2) 
# use gettext and strip methods to find each thread element in the 1 st table. 
header3 = [th.getText().strip() for th in data_tables[2][0].findAll('thead')[0].findAll('th')]
# print headers. 
header3
```

```python
# In the 3rd data tablke find all body elements and find all tr elements and assign to rows 3 
rows3 = data_tables[2][0].findAll('tbody')[0].findAll('tr')
```

```python
#find all of the td elements in rows 3 and assign to data_rows3
# itterate through each tr element in rows3
#for each td element found call find the right text passsing and index an the td element. 
data_rows3 = [[find_right_text(i, td) for i, td in enumerate(tr.findAll('td'))] for tr in rows3]
```

```python
# Build data frme with data from data_rows3 and column names from header 3
df3 = pd.DataFrame(data_rows3, columns=header3)
# display source name and table 
# use index 2
print(source_names[2])
df3.head()
```

```python
# 1) read in visit_data.csv file 

# Define the file path to CSV file 
file_path = 'visit_data.csv'

# use read_csv to read the file into a DataFrame. 
df = pd.read_csv(file_path)

# preview using head method. 

df.head()
```

```python
# 2) Check for duplicates. 

# Filter df to only include duplicate rows. 
duplicate_rows = df[df.duplicated()]
# display shape. 
duplicate_rows.shape
```

Here we can see that there are no completer repeates or duplicates of entire rows. 

```python
# 3) Check if any essential column contains NaN. 
# The essential columns would be email, IP and # visits. 

# Use the is null and any methods to determin if any value in an essential column is null. 
print('Are any emails null: ', (df['email'].isnull().any()))
print('Are any IPs null: ', (df['ip_address'].isnull().any()))
print('Are any visits null: ', (df['visit'].isnull().any()))
```

Here we can see that some visits are null or contatin NaN

```python
# 3.5 record current size of the df 

# use shape to store original size of df

original_size = df.shape

print('Original Size of DataFrame: ', original_size)
```

```python
#4) Get rid of outliers. 
```

```python
# Insure that all number of visits are finite value and not a decimal. 

# Build function to determine if each value is an intiger 
def is_integer(value):
    # Check that value is an intiger or a float that represents a whole number. 
    return isinstance(value, (int, float)) and value.is_integer()

# Loop through the column and remove rows where the value is not an integer
df_finite = df[df['visit'].apply(is_integer)]

# assign size after removing decimal visits.
size_after_int = df_finite.shape

print('Size of DataFrame after removing non finite values: ', size_after_int)
```

```python
# 5) report the size difference. 

print('The difference in the size of the data sets is: ', 
      (original_size[0]- size_after_int[0]),' Less rows.')

# print original and new size. 
print('The original size of the data: ', original_size[0])
print('The new size of the data is: ', size_after_int[0])
```

```python
# 6 Create a box plot to check for outliers 

# use matplot lib to create box plot. 
# title
plt.title('visits boxplot')
plt.boxplot(df_finite.visit)

#show
plt.show()
```

```python
# 7) get rid of any outliers. 
```

Based on the boxplot above we can detremine that any visit values above 2900 and below 100 are outliers. 

```python
# Filter dataset ro remove outliers 

# Filter for data above 100 and below 2900.
df_box = df_finite[(df_finite['visit'] <= 2900) & (df_finite['visit'] >= 100)]

# Determing the size of new df. 
size_after_box = df_box.shape

# Print difference 
print('This step has removed: ' , (size_after_int[0]-size_after_box[0]), 'rows') 

#print the new size of the dataset. 
print('The new size of the data set after removing outliers from boxplot is: ', size_after_box[0])
```

```python
# For fun lets remove outliers by z-score. 
```

```python
# Import packages. 
from scipy import stats
```

```python
# initialize new df
df_z_score = df_box

# calculate the z score of visits. add to df. 
df_z_score['z-score'] = stats.zscore(df_box.visit)


```

```python
# remove rows with z score above 3 
df_z_score_no_outliers = df_z_score[(df_z_score['z-score'] < 3)]
```

```python
# display changes to df

print('The new data set after removing z score outliers: ', df_z_score_no_outliers.shape[0])
```

removing outliers by z-score did not effect out dataset size.
