# Transformation 1: Remove dagger mark from "Name" column in wiki DataFrame. 

This will ensure that we can join the two DataFrames on the "Name" and "Ballpark" columns respectively. Ensuring that we can join the two DataFrames into one. 

```python
# Use the str.replace method to remove the dagger character from every value in the "Name" column. 
# Use wiki dataset
wiki['Name'] = wiki['Name'].str.replace('‡', '')
```

# Transformation 2: Eliminate Non-Active ballparks in Steam Dataset. 

This will ensure that we are merging our data down the line properly and referring to the proper ballpark. For example, Yankee Stadium has multiple listings in the steam dataset as there have been multiple Yankee Stadiums historically. 

```python
# Use the isnull() method to filter the steam dataset for ballparks where the last game is null.
# This will return only active ballparks. 
steam = steam[steam['Last Gm'].isnull()]
```

```python
# Display the resulting DataFrame. 
steam
```

Here we see that we have successfully removed all non-active ballparks from the steam DataFrame. There are more than 30 rows is this DataFrame because there are additional neutral site ballparks included in this DataFrame.

# Transformation 3: Perform Fuzzy Matching on "Name" and "Ballpark" 

Performing the fuzzy matching between the two keys from the wiki and steam DataFrames will allow us to merge the DataFrames into one in a later step. To do this we need to create matching keys in each DataFrame. 

```python
# Import Packages. 
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
# Apply fuzzy matching to create matching name column in each df. 

# Use the process.extractOne method from fuzzywuzzy and token_sort_ratio to find the best match for each name in a given DataFrame. 
# use the apply method to assign the matched name to the new column 'matched_name'
wiki['matched_name'] = wiki['Name'].apply(lambda x: process.extractOne(x, steam['Ballpark'], scorer=fuzz.token_sort_ratio)[0])
steam['matched_name'] = steam['Ballpark'].apply(lambda x: process.extractOne(x, wiki['Name'], scorer=fuzz.token_sort_ratio)[0])
```

```python
# Verify the results by printing a section of wiki. 

# Head wiki
wiki
```

Here we can see that everything matched well except the matched value for Angel Stadium. Lets Correct that by hand. 

```python
# Use the Loc method to replace the matched value for Angel stadium. 
# Index 1
# Steam name = 'Angel Stadium of Anaheim'
wiki.loc[1, 'matched_name'] = 'Angel Stadium of Anaheim'

# verify the results . 
# Display first 5 rows. 
wiki.head(5)
```

Now we have successfully completed a fuzzy matching. This work will allow us to move on to our next transformation and join the two DataFrames together. 

# Transformation 4: Merge both the wiki and steam DataFrames together. 

Here we will use a right outer join in order to add the information from the steam DataFrame to the wiki DataFrame. This join will be done on "Ballpark" and "matched_name" (from the wiki DataFrame) respectively. 

```python
# Use merge method to join both datasets.
# Use right outer join.
# Keys: wiki = "matched_name", steam = "Ballpark". 
ballpark = wiki.merge(steam, how='inner', left_on='matched_name', right_on='Ballpark')
```

```python
# Display the new dataset 
ballpark
```

We have successfully completed a join of the two separate DataFrames from our websites Wikipedia and Steamheads. Now working with one file we can begin to clean the DataFrame to make it more workable. Additionally we know that we have accounded for information from all 30  active Major League ballparks.
