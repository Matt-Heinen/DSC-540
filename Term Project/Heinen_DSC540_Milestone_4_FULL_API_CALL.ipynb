# Read in Data From Open Weather 3.0 API

Here we will read in data from the Open Weather API. This will allow us to create a dataset that has the stored weather data from each ball put in play in Carlos Santana's career. 

```python
import pandas as pd
import requests


# Define a function to make the API call and return the JSON data
def call_api(row):
    url = 'https://api.openweathermap.org/data/3.0/onecall/timemachine?'
    params = {
        'lat': row['Lat'],
        'lon': row['Long'],
        'dt': row['unix_time'],
        'appid': '6b1915f8a711e908db855ea4824218f5'
        
    }
    response = requests.get(url, params=params)
    return response.json()
```

```python
# Create an empty list to store the returned JSON data
json_data_list = []
```

```python
# Loop through the DataFrame, call the API for each row, and store the returned JSON data
for index, row in carlos_lo.iterrows():
    json_data = call_api(row)
    json_data = call_api(row)
    json_data_list.append(json_data)


# Now result_df contains the JSON data returned from the API calls
```

```python
# Create a new DataFrame from the list of JSON data
carlos_weather = pd.DataFrame(json_data_list)

# Preview 
carlos_weather.head(10)
```

# Transformation 4: Convert Temperatures to Fahrenheit. 

This will store our temperature data in a way that is readable and understandable for our viewers.

The conversion formula from kelvin to Fahrenheit is shown below: 

         Â°F = (K âˆ’ 273.15) Ã— (9/5) + 32

We will apply this formula to the temp, dew_point and feels_like columns. 

```python
# Define a function to convert Kelvin to Fahrenheit
def kelvin_to_fahrenheit(k):
    return k * 9/5 - 459.67
```

```python
# Apply the conversion function kelvin_to_fahrenheit to temp 
carlos_weather['temp'] = carlos_weather['temp'].apply(kelvin_to_fahrenheit)
```

```python
# Apply the conversion function kelvin_to_fahrenheit to feels like. 
carlos_weather['feels_like'] = carlos_weather['feels_like'].apply(kelvin_to_fahrenheit)
```

```python
# Apply the conversion function kelvin_to_fahrenheit to dew_point
carlos_weather['dew_point'] = carlos_weather['dew_point'].apply(kelvin_to_fahrenheit)
```
